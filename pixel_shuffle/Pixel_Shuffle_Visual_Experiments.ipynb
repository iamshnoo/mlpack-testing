### Imports


````
%%capture
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
````

### PyTorch official example

```python

# Example given in PyTorch docs:

pixel_shuffle = nn.PixelShuffle(3)
input = torch.randn(1, 9, 4, 4)
output = pixel_shuffle(input)
print(output.size()) # torch.Size([1, 1, 12, 12])

```

### Experiment Set 1

````
input = torch.zeros([1, 9, 28, 28])
v = np.arange(1,28*28 + 1)
v = v.reshape(28,28)
input[0][0] = torch.from_numpy(v) 
pixel_shuffle = nn.PixelShuffle(3)
out = pixel_shuffle(input) # torch.size([1, 1, 84, 84])

num_cols = input.shape[1]+1
num_rows = input.shape[0]
fig = plt.figure(figsize=(40,40))
input = input.squeeze()
for i in range(10):
    ax1 = fig.add_subplot(num_rows, num_cols, i+1)
    if i==9:
      x = out.squeeze()
      if (x.ndim==3):
        x = np.transpose(x, (1,2,0))
      print('Each output channel is of shape {}\n'.format(x.shape))
      ax1.imshow(x)
    else:
      temp = input[i].unsqueeze(0)
      temp = np.transpose(temp, (1,2,0))
      temp = temp.squeeze().numpy()
      if i==0:
        print('Each input channel is of shape {}\n'.format(temp.shape))
      ax1.imshow(temp)

    ax1.axis('off')
    ax1.set_xticklabels([])
    ax1.set_yticklabels([])

fig.show()
fig.savefig('1.png', dpi=100,  bbox_inches='tight')
````
